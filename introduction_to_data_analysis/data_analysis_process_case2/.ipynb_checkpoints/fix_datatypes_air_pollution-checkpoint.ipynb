```python
# load datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('darkgrid')

alpha08 = pd.read_csv('data_08_v3.csv', sep = ';')
alpha18 = pd.read_csv('data_18_v3.csv', sep = ';')
```

```python
# try using pandas' to_numeric or astype function to convert the
# 2008 air_pollution_score column to float -- this won't work

alpha08.air_pollution_score = alpha08.air_pollution_score.astype(float)
```

# Figuring out the issue

```python
alpha08[alpha08.air_pollution_score == '6/4']
```

# It's not just the air pollution score!

```python
# First, let's get all the hybrids in 2008
hb_08 = alpha08[alpha08['fuel'].str.contains('/')]
hb_08
```

```python
# hybrids in 2018
hb_18 = alpha18[alpha18['fuel'].str.contains('/')]
hb_18
```

```python
# create two copies of the 2008 hybrids dataframe
df1 = hb_08.copy()  # data on first fuel type of each hybrid vehicle
df2 = hb_08.copy()  # data on second fuel type of each hybrid vehicle

# Each one should look like this
df1
```

```python
# columns to split by "/"
split_columns = ['fuel', 'air_pollution_score', 'city_mpg', 'hwy_mpg', 'cmb_mpg', 'greenhouse_gas_score']

# apply split function to each column of each dataframe copy
for c in split_columns:
    df1[c] = df1[c].apply(lambda x: x.split("/")[0])
    df2[c] = df2[c].apply(lambda x: x.split("/")[1])
```

```python
# this dataframe holds info for the FIRST fuel type of the hybrid
# aka the values before the "/"s
df1
```

```python
# this dataframe holds info for the SECOND fuel type of the hybrid
# aka the values after the "/"s
df2
```

```python
# combine dataframes to add to the original dataframe
new_rows = df1.append(df2)

# now we have separate rows for each fuel type of each vehicle!
new_rows
```

```python
# drop the original hybrid rows
alpha08.drop(hb_08.index, inplace=True)

# add in our newly separated rows
alpha08 = alpha08.append(new_rows, ignore_index=True)
```

```python
# check that all the original hybrid rows with "/"s are gone
alpha08[alpha08['fuel'].str.contains('/')]
```

```python
alpha08.shape
```

# Repeat this process for the 2018 dataset

```python
# create two copies of the 2018 hybrids dataframe, hb_18
df1 = hb_18.copy()
df2 = hb_18.copy()
```

```python
# list of columns to split
split_columns = ['fuel', 'city_mpg', 'hwy_mpg', 'cmb_mpg']

# apply split function to each column of each dataframe copy
for c in split_columns:
    df1[c] = df1[c].apply(lambda x: x.split("/")[0])
    df2[c] = df2[c].apply(lambda x: x.split("/")[1])
```

```python
# append the two dataframes
new_rows = df1.append(df2)

# drop each hybrid row from the original 2018 dataframe
# do this by using pandas' drop function with hb_18's index
alpha18.drop(hb_18.index, inplace=True)

# append new_rows to df_18
alpha18 = alpha18.append(new_rows, ignore_index=True)
```

```python
# check that they're gone
alpha18[alpha18['fuel'].str.contains('/')]
```

```python
alpha18.shape
```

### Now we can comfortably continue the changes needed for `air_pollution_score`! Here they are again:
- 2008: convert string to float
- 2018: convert int to float

```python
# convert string to float for 2008 air pollution column
alpha08.air_pollution_score = alpha08.air_pollution_score.astype(float)
```

```python
# convert int to float for 2018 air pollution column
alpha18.air_pollution_score = alpha18.air_pollution_score.astype(float)
```

```python
alpha08.to_csv('data_08_v4.csv', index=False, sep = ';')
alpha18.to_csv('data_18_v4.csv', index=False, sep = ';')
```
