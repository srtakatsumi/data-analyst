```python
# load datasets

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('darkgrid')

alpha08 = pd.read_csv('clean_08.csv', sep = ';')
alpha18 = pd.read_csv('clean_18.csv', sep = ';')
```

```python
alpha08.shape
```

```python
alpha18.shape
```

### Create combined dataset

```python
# rename 2008 columns
alpha08.rename(columns=lambda x: x[:20] + "_2008", inplace=True)
```

```python
alpha18.rename(columns=lambda x: x[:20] + "_2018", inplace=True)
```

```python
# view to check names
alpha08.head()
```

```python
alpha18.head()
```

```python
# merge datasets
df_combined =alpha08.merge(alpha18, left_on='model_2008', right_on='model_2018', how='inner')
```

```python
# view to check merge
df_combined.head()
```

```python
df_combined.shape
```

```python
df_combined.to_csv('combined_dataset.csv', index=False, sep = ';')
```
